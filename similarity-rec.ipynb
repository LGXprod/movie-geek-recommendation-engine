```python
import json
import pandas as pd
import numpy as np
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
ratings_matrix = np.load('ratings_matrix.npy')
```

```python
with open('movie_ids_int.json', 'r') as fp:
    movie_ids_to_index = json.load(fp)

with open('user_ids.json', 'r') as fp:
    user_ids = json.load(fp)
```

```python
movie_ids_to_index
```

```python
# iron man, iron man 2, capital america the 1st avenger, hulk, avengers (2012)
movie_ids = [59315, 77561, 88140, 6534, 89745]
movie_indices = [movie_ids_to_index[str(movie_id)] for movie_id in movie_ids]
movie_ratings = [5, 4, 4, 4, 5]
user_movie_ratings = np.zeros((26744,), dtype=np.float32)

for i, movie_index in enumerate(movie_indices):
    user_movie_ratings[movie_index] = movie_ratings[i]
```

```python
movie_indices
```

```python
user_movie_ratings[1001]
```

```python
similarity_scores = []

for i, other_user_movie_ratings in enumerate(ratings_matrix):
    a = tf.constant(user_movie_ratings, dtype=tf.float32)
    b = tf.constant(other_user_movie_ratings, dtype=tf.float32)
    s = tf.losses.cosine_distance(tf.nn.l2_normalize(a, 0), tf.nn.l2_normalize(b, 0), dim=0)
    similarity = 1 - tf.Session().run(s)
    similarity_scores.append(similarity)
    print(f"User: {i}", end="\r")
```
